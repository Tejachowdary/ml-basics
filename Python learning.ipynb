# Python OOPS- Public, Protected And Private


```python
### public class variables can be accessed and modified either in the object or in the child class
### protected variables(declared by _variablename) should be ideally modified only in the child class
            #### in dir of object, attributes would be as _attributename
### private variables(declared by __variablename)should not be ideally modified anywhere
            #### in dir of object, attributes would be as _classname__attributename

## Python is linient to allow modifying any variable irrespective of whether it is Public, Protected And Private
```

```python
## protected variables
class vehicleProtected:
    def __init__(self,windows,engineType):
        self._windows=windows
        self._engineType=engineType

        
```

```python
vehicle1=vehicleProtected(2,"diesel")
```

```python
dir(vehicle1)
```

```python
class childVehicle(vehicleProtected):
    def __init__(self,windows,engineType,sunroof):
        super().__init__(windows,engineType)
        self.sunroof=sunroof
```

```python
childVehicleObject=childVehicle(2,"diesel",True)
```

# z-score statistics- To convert guassian distir to standard normal distirbution
normal/guassian distirbution-center-mean
                                one std right-mean+std
                                one std left-mean-std
               data distirbution in mean-std,mean,mean+std is 68%
                                    mean-2std,mean-std,mean,mean+std,mean+2std is 95%
                                   entire-mean-3std, mean-2std,mean-std,mean,mean+std,mean+2std,mean+3std is 99%
               
                                
standard normal distirbution(snd)-mean=0,std=1
                                 one std right-mean+std=0+1=1
                                one std left-mean-std =0-1=-1
 Z-score=xi-mean/std
 In Feature engineering we apply this to a feature .It is called Normalization/standard normalization
                                
Problem:
  mean=75,std=10
  Find p(x>60)
Ans:z=60-75/10= -1.5
BY finding value from z score table for -1.5 we get 0.0668
p(x>60)=0.5-0.0668+0.5=0.93332

# # Finding outliers using Z score

```python
# Finding outliers using python
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Finding outlier using Z score=xi-mean/std
outlier=[]
def detect_outlier(data):
    mean=np.mean(data)
    std=np.std(data)
    #data over third std 
    threshold=3
    
    for i in data:
        zScore=(i-mean)/std
        if(np.abs(zScore)>threshold):
            outlier.append(i)
    return outlier
```

```python
detect_outlier(dataset)
```

# T test
It is type of inferential stats which is used to determine if there is a significant difference between the maens of two groups which may be related in certain feature

one sample t test

two sample t tests

One sample t test- It will tell us if means of sample and population are different are not 

```python
# example1
# population 
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
import numpy as np;
# pop mean
np.mean(ages)

```

```python
#sample of size 10 and calculate its mean
age_sample=np.random.choice(ages,10)
```

```python
from scipy.stats import ttest_1samp
#ttest_isamp(sample, Poulation mean)-> it returns p value
TtestStat,pValue=ttest_1samp(age_sample,30)
```

```python
if pValue>0.05:
    print("Accept null hypothesis")
else:
    print("Reject null hypothesis")
```

# Two sample t test
compares the means of two independent group inorder to determine whether there is a statistical evidence that the associated poulation means are significantly different. Also known as independent t test

```python
# population 1
population1=stats.poisson.rvs(loc=18,mu=33,size=60)

# population 2
population2=stats.poisson.rvs(loc=11,mu=33,size=70)

```

```python
from scipy.stats import ttest_ind
_,p_value=ttest_ind(population1,population2)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

# Chi squared test - two categorical variables

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
# relation b/n sex and smoker
data_table=pd.crosstab(data.sex,data.smoker)
observed_value=data_table.values
chi2_stat, p_value,dof,expected_value=stats.chi2_contingency(data_table)
```

```python
#Approach 1 using scipy library
if(p_value<=0.05):
    print("Reject null hypothesis(BOth fetaures are not independent)")
else:
    print("Accept Null hypothesis(Both fetaures are independent)")
```

```python
observed_value
```

```python
expected_value
```

```python
#Approach 2 manual calc
#ddof=no of rows-1*no of coulumns -1
data_table_shape=data_table.shape
ddof=(data_table_shape[0]-1)*(data_table_shape[1]-1)
another_approach_p_value=1-stats.chi2.cdf(chi2_stat,ddof)
#p_value and another_approach_p_value are same

chi_square_statistic=sum(sum([(o-e)**2./e for o,e in zip(observed_value,expected_value)]))

critical_value=stats.chi2.ppf(q=1-0.05,df=ddof)
if(critical_value<=chi_square_statistic):
    print("reject null hypothesis-features are not independent")
else:
    print("accept null hypothesis-features are independent")
```

# Logistic regression-
Regression-numeric valued output
    linear regression- continuous numeric valued output- e.g. price of house
    logistic regression-discrete numeric valued output- e.g pass/fail in test
    Both linear and logistic regression learns from data exaples-supervised learning
    
    Linear reg is based on selecting a best fit line which minimises sum of distance b/n actual points and predicted points
    Linear reg cannot be used for finding discrete valued output(classification) as Presenece of outliers gives bad best fit line for data and output can be of any value(Not fixed discrete values)
    
    Logistic regression solves this problem by using maximizing summation for all points of sigmoid function(only two outputs 0 or 1 for any input - for fixing disrete values) & (Minimises impact of outliers by scaling it down( to 0 or 1)) of yi*(theta transpose).xi for different theta values to find the theta which provides best fit line that lineraly seperates classes in classification problem
